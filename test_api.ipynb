```python
import requests
```

```python
url = "http://127.0.0.1:5000/predict"

data = {
    'age': 25,
    'gender' : 'male',
    'subscription_type' : 'Standard',
    'last_login_days': 45,
    'region' :'Goa',
    'device' : 'Laptop',
    'monthly_fee': 499,
    'payment_method': 'UPI',
    'number_of_profiles': 4,
    'avg_watch_time_per_day': 3.4,
    'favorite_genre': "Romance" 
}

print("Sending request to:", url)
print("Request data:", data)
print("-" * 50)

try:
    response = requests.post(url, json=data, timeout=5)
    print(f"Status Code: {response.status_code}")
    print(f"Response Headers: {response.headers}")
    print(f"Response Text: {response.text}")
    
    if response.status_code == 200:
        result = response.json()
        print("\n✓ SUCCESS - Prediction Results:")
        print(f"  Prediction: {result['Prediction']}")
        print(f"  Churn Probability: {result['Churn Probability']}")
        print(f"  Retention Strategy: {result['Retention Strategy']}")
    else:
        print(f"\n✗ Error: {response.status_code}")
        if response.text:
            print(f"  Details: {response.text}")
        
except requests.exceptions.ConnectionError:
    print("✗ Connection Error: Cannot connect to http://127.0.0.1:5000")
    print("  Make sure the Flask app is running!")
except requests.exceptions.Timeout:
    print("✗ Timeout Error: Request took too long")
except Exception as e:
    print(f"✗ Error: {type(e).__name__}: {e}")
```
